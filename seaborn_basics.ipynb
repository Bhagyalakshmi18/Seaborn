## Importing required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## I. Statistical Relationships
### 1. Scatter Plot

```python
a = sns.load_dataset("flights")
sns.relplot(x = "passengers", y="month", hue="year", data=a)
```

### 2. Line Plot

```python
b = sns.load_dataset("tips")
sns.relplot(x = "time", y="tip", kind="line", data=b)
```

## II. Categorical Data

```python
sns.catplot(x="day", y="total_bill", data=b)
```

### Violin plot

```python
sns.catplot(x="day", y="total_bill", data=b, kind="violin")
```

### Boxen plot

```python
sns.catplot(x="day", y="total_bill", data=b, kind="boxen")
```

## III. Visualizing distribution of dataset
### 1. Univariate distribution

```python
c = np.random.normal(loc=5, size=100, scale=2)
sns.distplot(c)
```

## 2. Bivariate distribution (Joint plot)
#### We can only use kind: scatter, reg, resid, kde, hex
#### axes_style : white, dark, whitegrid, darkgrid, ticks
#### Multivariate_normal : for assembling the graph in different shapes and normal for assembling graph in single shape

```python
x = pd.DataFrame({'Day':[1,2,3,4,5,6,7], 'Grocery':[38,57,34,79,78,29,27], 'Clothes':[39,68,92,37,48,93,82], 'Utensils':[18,89,27,49,92,28,21]})
y = pd.DataFrame({'Day':[8,9,10,11,12,13,14], 'Grocery':[38,57,34,79,78,29,27], 'Clothes':[39,68,92,37,48,93,82], 'Utensils':[18,89,27,49,92,28,21]})
mean, cov = [0,1], [(1,0.5), (0.5,1)]
data = np.random.multivariate_normal(mean,cov,100)
with sns.axes_style("dark"):
    sns.jointplot(x=x, y=y, kind="kde", color="b");
```

## Multi-Plot grids
### 1. Facet Grid

```python
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

### 2. Pair Grid

```python
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

### Plot Aesthetics

```python
sns.set(style="darkgrid")
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

## Reading numerical data and plotting graphs

```python
data_ = pd.read_csv("Marketing Raw Data.csv")
sns.lineplot(x = "Date", y="Revenue", data=data_)
```

#### Increasing the size of graph using figsize, ci stands for confidence interval

```python
sns.set(rc={'figure.figsize':(12,10)})
sns.lineplot(x = "Week_ID", y="Revenue", data=data_, hue="Promo", style="Promo", ci=False, markers=True)
```

```python
sns.barplot(x = "Month_ID", y="Revenue", data=data_, ci=None, hue="Promo")
data_[['Revenue','Month_ID']].groupby('Month_ID').agg({'Revenue':'sum'})
```

```python
sns.barplot(x = "Revenue", y="Month_ID", data=data_, ci=None, hue="Promo", orient="h", color="green")
```
